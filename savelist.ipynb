```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import json

#Json  
origin = pd.read_json('./3_IRT.json')

print(len(origin))
print("=========================")
print("start separation", end="\n\n")

# dataframe 
m1 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])
m2 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])
m3 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])

for idx, row in origin.iterrows():

    profile = row.learnerProfile
    if(profile == "F;S01;7" or profile == "M;S01;7"):
        m1 = m1.append(row, ignore_index=True)
    elif(profile == "F;S01;8" or profile == "M;S01;8"):
        m2 = m2.append(row, ignore_index=True)
    elif(profile == "F;S01;9" or profile == "M;S01;9"):
        m3 = m3.append(row, ignore_index=True)

    if (idx % 300 == 0):
        print(idx)
    elif (idx % 10 == 0):
        print(".", end='')



print("=========================")
#print(m1)
#print(m2)
#print(m3)


#  ,   
matplotlib.rcParams['axes.unicode_minus'] = False

# 0.4  
m1 = m1[m1['realScore'] <= 0.4]
m2 = m2[m2['realScore'] <= 0.4]
m3 = m3[m3['realScore'] <= 0.4]

m_1 = m1.to_json(orient = 'records')
m_2 = m2.to_json(orient = 'records')
m_3 = m3.to_json(orient = 'records')

with open('m1_filter_realscore.txt', 'w') as outfile:
    outfile.write(m_1)
with open('m2_filter_realscore.txt', 'w') as outfile:
    outfile.write(m_2)
with open('m3_filter_realscore.txt', 'w') as outfile:
    outfile.write(m_3)

    
```
