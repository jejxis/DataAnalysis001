```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Json  
origin = pd.read_json('./3_IRT.json')

print(len(origin))
print("=========================")
print("start separation", end="\n\n")

# dataframe 
m1 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])
m2 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])
m3 = pd.DataFrame(columns=['learnerID', 'learnerProfile', 'testID', 'theta', 'realScore', 'TimeStamp'])

for idx, row in origin.iterrows():

    profile = row.learnerProfile
    if(profile == "F;S01;7" or profile == "M;S01;7"):
        m1 = m1.append(row, ignore_index=True)
    elif(profile == "F;S01;8" or profile == "M;S01;8"):
        m2 = m2.append(row, ignore_index=True)
    elif(profile == "F;S01;9" or profile == "M;S01;9"):
        m3 = m3.append(row, ignore_index=True)

    if (idx % 300 == 0):
        print(idx)
    elif (idx % 10 == 0):
        print(".", end='')



print("=========================")
#print(m1)
#print(m2)
#print(m3)


#  ,   
matplotlib.rcParams['axes.unicode_minus'] = False

#1 
plt.hist(m1['theta'], bins=20)
plt.xlabel("m1_theta")
plt.ylabel("count")

plt.grid()
plt.show()

#1 

plt.hist(m1['realScore'], bins=20)
plt.xlabel("m1_realScore")
plt.ylabel("count")

plt.grid()
plt.show()

#2 
plt.hist(m2['theta'], bins=20)
plt.xlabel("m2_theta")
plt.ylabel("count")

plt.grid()
plt.show()

#2 

plt.hist(m2['realScore'], bins=20)
plt.xlabel("m2_realScore")
plt.ylabel("count")

plt.grid()
plt.show()

#3 
plt.hist(m3['theta'], bins=20)
plt.xlabel("m3_theta")
plt.ylabel("count")

plt.grid()
plt.show()

#3 

plt.hist(m3['realScore'], bins=20)
plt.xlabel("m3_realScore")
plt.ylabel("count")

plt.grid()
plt.show()
```
